```python
def universal_acclaim_filter(df, col1, col2, avg_col, diff_threshold, avg_threshold):
    """
    Filters rows where the absolute difference between col1 and col2 is less than diff_threshold,
    and avg_col is greater than avg_threshold.
    Returns the filtered dataframe sorted by index.
    """
    return df[(abs(df[col1] - df[col2]) < diff_threshold) & (df[avg_col] > avg_threshold)].sort_index()
```

```python
UniversalAcclaim_jobs = universal_acclaim_filter(
    jobs,
    col1='seniority_num',
    col2='months_since_posted',
    avg_col='Avg_Experience_Freshness',
    diff_threshold=3,
    avg_threshold=4
)

print(UniversalAcclaim_jobs.head())
```

```python
#Demographic analysis
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv(r"C:\Users\User\jupiternotebooks\1000_ml_jobs_us.csv")

# Step 3: Check basic info
print("Shape of dataset:", df.shape)
print("Columns:\n", df.columns)
df.head()
```

```python
# Step 4: Create a DataFrame similar to df_by_genre
# We'll use job_title and seniority_level as categorical features
df_by_role = df[['job_title', 'seniority_level']].copy()

# Add a count column (like 'cnt' in the IMDB tutorial)
df_by_role['cnt'] = 1

df_by_role.head()
```

```python
# Step 5: Group by job_title
df_by_job_title = df_by_role.groupby('job_title').sum(numeric_only=True).sort_values('cnt', ascending=False)

# Display top 10 job titles
print("Top Job Titles by Count:\n")
print(df_by_job_title.head(10))
```

```python
# Step 6: Group by seniority_level
df_by_seniority = df_by_role.groupby('seniority_level').sum(numeric_only=True).sort_values('cnt', ascending=False)

# Display top 10 seniority levels
print("Top Seniority Levels by Count:\n")
print(df_by_seniority.head(10))
```

```python
# Step 7: Visualize job titles
plt.figure(figsize=(10, 6))
df_by_job_title.head(10).plot(kind='barh', legend=False, title='Top 10 Job Titles by Count', color='skyblue')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.gca().invert_yaxis()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#split in  dataset
# Mark rows as 'USA' or 'Non-USA' based on 'company_address_region' values

# Example: if 'company_address_region' contains US state abbreviations or full names, mark as 'USA', else 'Non-USA'

# Let's create a list of US states abbreviations or full names (simplified)
us_states = [
    'CA', 'TX', 'NY', 'FL', 'IL', 'PA', 'OH', 'GA', 'NC', 'MI', 
    'New York', 'California', 'Texas', 'Florida', 'Illinois', 'Pennsylvania', 'Ohio', 'Georgia', 'North Carolina', 'Michigan'
]

def mark_usa_nonusa(region):
    if pd.isna(region):
        return 'Unknown'
    return 'USA' if region in us_states else 'Non-USA'

df['usa_or_not'] = df['company_address_region'].apply(mark_usa_nonusa)
```

```python
df['job_description_length'] = df['job_description_text'].apply(lambda x: len(str(x)))
```
